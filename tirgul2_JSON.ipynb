```python
import json
import ipywidgets as widgets
from IPython.display import display, clear_output

# Path to the JSON file
path = '/content/drive/My Drive/Colab Notebooks/Cloud Braude/students.json'

# Load student data
with open(path, 'r') as fid:
    students = json.load(fid)

# Create list of student names for selection
student_names = [s['name'] for s in students]

# Dropdown menu for selecting a student
dropdown = widgets.Dropdown(
    options=student_names,
    description='Student:',
    style={'description_width': 'initial'}
)

# Text boxes for student details
surname_box = widgets.Text(description='Surname:')
email_box = widgets.Text(description='Email:')
courses_box = widgets.Text(description='Courses:')
link_box = widgets.Text(description='Link:')
age_box = widgets.Text(description='Age:')

# Update button
update_button = widgets.Button(description='Update Age', button_style='success')

# Output display
output = widgets.Output()

# Function to display selected student info
def on_student_change(change):
    clear_output(wait=True)
    selected_name = change['new']
    student = next(s for s in students if s['name'] == selected_name)
    surname_box.value = student['surname']
    email_box.value = student['email']
    courses_box.value = ', '.join(student['courses'])
    link_box.value = student['link']
    age_box.value = str(student.get('age', ''))  # default if age not set
    display(dropdown, surname_box, email_box, courses_box, link_box, age_box, update_button, output)

# Function to update student age and save to file
def update_age(b):
    selected_name = dropdown.value
    for student in students:
        if student['name'] == selected_name:
            student['age'] = age_box.value
    with open(path, 'w') as fid:
        json.dump(students, fid, indent=4)
    with output:
        clear_output()
        print(f"Age for {selected_name} updated to {age_box.value} and saved to file!")

# Connect events
dropdown.observe(on_student_change, names='value')
update_button.on_click(update_age)

# Initial display
display(dropdown)
```
